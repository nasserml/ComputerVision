```python
import cv2
```

```python
print(cv2.__version__)
```

```python
img = cv2.imread('apple.jpg', 0)

print(img)
```

```python
cv2.imshow('first image', img)

k = cv2.waitKey(0)

if k == 27 :
    cv2.destroyAllWindows()
elif k == ord('s'):
    cv2.imwrite('sec image.jpg', img)
    cv2.destroyAllWindows()
```
